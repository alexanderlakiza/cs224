**Индексы**




> *Индекс* - номер символа в строке.


```
строка[индекс]
```


Символы могут быть пронумерованы слева направо - от `0`. Последний символ: `длина строки - 1`
Символы могут быть пронумерованы справа налево - от `-1`. Первый символ: `-длина строки`


````
stroka = 'story'
print(stroka[0])
print(stroka[-len(stroka)])
print(stroka[len(stroka) - 1])
print(stroka[-1])
````

````
print(stroka[2])
````

Нельзя обратиться к индексу, которого нет.

````
print(stroka[5])
````

Нельзя заменить символ, обратившись к нему по индексу.

````
stroka[2] = 'a'
````

**Срезы**



```
строка[номер_первого_элемента:номер_следующего_за_последним_элементом]
```

````
greeting = 'Hello World!'

print(greeting[6:11])
````

````
print(greeting[:5])
````

````
print(greeting[6:])
````

````
print(greeting[:])
````

````
print(greeting[-5:-2])
````

````
print(greeting[-5:])
````

````
print(greeting[:-2])
````

````
print("|", greeting[1:6:2], "|")
````

````
print(greeting[9:2:-3])
````

````
print(greeting[::-1])
````

````
for i in range(len(stroka)):
  if i != len(stroka) - 1:
    print(stroka[i] + stroka[i + 1])
````

````
for chr in stroka:
  print(chr)
````

**МЕТОДЫ ДЛЯ РАБОТЫ СО СТРОКАМИ**



> **Метод** - как функция, но вызывается после объекта через точку. Вот так:


```
объект.метод()
```





**`1. index(подстрока, [начало], [конец])`**

Позволяет определить местоположение заданной подстроки в исходной строке, т.е. проверяет, есть ли подстрока в строке, и если есть - выводит индекс ее первого символа.

````
greeting = 'Hello World!'
print(greeting.index('World'))
````

````
print(greeting.index('l'))
````

Можно указать диапазон, в котором нужно найти подстроку.

````
print(greeting.index('l', 3, 9))
````

Если нужного символа/подстроки в строке нет, `Python` выдаст ошибку.

````
print(greeting.index('s'))
````

````
a = "lsfknv"
print(sorted(a))
````

**`2. rindex(подстрока, [начало], [конец])`**

Как `index()`, только выводит индекс последнего найденного символа.

````
greeting = 'Hello World World!'
print(greeting.rindex('World'))
````

````
print(greeting.rindex('l'))
````

**`3. find(подстрока, [начало], [конец])`**

Ищет подстроку в строке. Возвращает номер первого вхождения или `-1`, если ничего не найдено.

````
greeting = 'Hello World!'
print(greeting.find('l'))
````

````
print(greeting.find('f'))
````

**`4. rfind(подстрока, [начало], [конец])`**

Ищет подстроку в строке. Возвращает номер последнего вхождения или `-1`, если ничего не найдено.

````
greeting = 'Hello World!'
print(greeting.rfind('o'))
````

````
print(greeting.rfind('lo'))
````

**`5. count()`**

Позволяет посчитать, сколько раз подстрока встречается в строке.

````
greeting = 'Hello Worldlo!'
print(greeting.count('l') + greeting.count('o'))
````

Метод считает только не перекрывающиеся строки.

````
letters = 'ababa'
print(letters.count('aba'))
````

**`6. split()`**

Разбивает строку на список подстрок, разделенных пробельными символами (пробел, табуляция, перенос строки).

````
numbers = 'One Two Three'
print(numbers.split())
````

````
numbers = 'One \n Two \n Three'
num_list = numbers.split()
print(num_list)
````

Можно задать разделитель самостоятельно. Например, запятую, двоеточие, любой другой знак препинания или букву.

````
numbers = 'One, Two,Three'
print(numbers.split(','))
````

**`7. join()`**

Объединяет список строк в одну строку.

````
print(', '.join(num_list))
````

**`8. replace()`**

Заменяет одну подстроку другой. При этом создается новая строка, а исходная не меняется.

````
greeting = 'Hello World!'
print(greeting.replace('l', 'U'))
````

Можно заменить не все вхождения, а только первые несколько.

````
print(greeting.replace('l', 'U', 1))
````

**`9. isdigit() и isalpha()`**

`isdigit()` проверяет, состоит ли строка полностью из цифр.

`isalpha()` проверяет, состоит ли строка полностью из букв.

````
greeting = 'Hello World'
print(greeting.isdigit())
````

````
print(greeting.isalpha())
````

````
greeting = 'абаваомыоарм'
print(greeting.isalpha())
````

````
s = "sdfvdfb"
print(s[1])
s[1] = "g"
print(s)
````

**`10. upper(), lower(), title()`**

`upper()` преобразует все буквы в строке в верхний регистр.

`lower()` преобразует все буквы в строке в нижний регистр.

`title()` преобразует первую букву каждого слова в верхний регистр, все остальные - в нижний.

````
greeting = 'Hello worLd!'
print(greeting.lower())
````

````
print(greeting.upper())
````

````
print(greeting.title())
````

Методы **`islower()`**, **`isupper()`**, **`istitle()`**, соответственно, проверяют, состоит ли строка только из букв нижнего регистра, верхнего или начинаются ли все слова с заглавной буквы. 

````
greeting = 'Hello World!'
print(greeting.islower())
````

````
print(greeting.isupper())
````

````
print(greeting.istitle())
````

**`11. startswith(подстрока) и endswith(подстрока)`**

Проверяют, начинается ли или заканчивается строка заданной подстрокой.

````
greeting = 'Hello World!'
print(greeting.startswith('He'))
````

````
print(greeting.startswith('he'))
````

````
print(greeting.endswith('!'))
````

**`12. capitalize()`**

Преобразует первый символ строки в верхний регистр, остальные - в нижний.

````
greeting = 'hello World!'
print(greeting.capitalize())
````

**`13. strip(), lstrip(), rstrip()`**

Удаляют пробелы в начале строки (`lstrip()`), в конце (`rstrip()`) или с обеих сторон (`strip()`).

````
greeting = '  Hello  World   '
print(greeting.strip())
````

````
print(greeting.lstrip())
````

````
print(greeting.rstrip())
````

**`14. swapcase()`**

Переводит символы нижнего регистра в верхний, а верхнего - в нижний.

````
greeting = 'Hello World!'
print(greeting.swapcase())
````

**`15. isspace()`**

Проверяет, состоит ли строка из пробельных символов.

````
greeting = 'Hello World!'
print(greeting.isspace())
````

````
print('\n'.isspace())
````

**`16. format()`**

Позволяет задать шаблон строки и указать, какие переменные нужно вставить в этот шаблон.

````
template = 'Число {} идет после {}.'
print('Число {} идет после {}.'.format(13, 12))
````

````
template = 'Число {1} идет после {0}. {0} меньше {1} на 1.'
print(template.format(12, 13))
````

**`f-строки`**

Позволяют вставить значение переменной прямо в строку, без использования метода `format()`.

````
x = 12
print(f"После {x} идет {x + 1}.")
````
