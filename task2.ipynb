### 2.1.1. Напишите регулярное выражение, которое возвращает список первых двух букв каждого слова строки. Обратите внимание на работу с дефисом.

Если мы **не показываем** слова из одной буквы:

```python
result = re.findall(r'\b\w\w', text)
result
```

Если мы **показываем слова** из одной буквы:

```python
re.findall(r'\b\w{1,2}', text)
```

Какие операторы использовались?

| Оператор | Что он делает |
|:-:|:-:|
| \b | граница слова |
| \w | любая цифра или буква |
| {n, m} | от n до m вхождений |

### 2.1.3. Напишите регулярное выражение, которое заменит все подстроки, обозначающие время (только время, не даты), в строке на 'TBD'.

```python
subtext1 = 'Он обещал приехать ровно в 14:31:58, но приедет не раньше 15:00'

pattern = r'((?:[01]\d|2[0-3])\:(?:[0-5]\d)(?:\:[0-5]\d)?)'
newsubtext1 = re.sub(pattern, "TBD", subtext1)

print(newsubtext1)
```

Какие операторы использовались?

| Оператор | Что он делает |
|:-:|:-:|
| \d | любая цифра [0-9] |
| [..] | один из символов в скобках |
| \ | экранирование спецсимволов |
| ? | 0 или 1 вхождение шаблона, стоящего перед ? |
| a\|b | равно a или b |

### 2.1.4. Напишите регулярное выражение, которое заменяет произвольное количество пробельных символов внутри строки на один пробел.

```python
subtext2 = "В чемпионате     мира по        автогонкам в серии Формула 1 уже  седьмой  год подряд выигрывает  немецкая        команда  Мерседес."
```

```python
print(subtext2)
```

```python
pattern = r'\s{1,}'

newsubtext2 = re.sub(pattern, ' ', subtext2)

print(newsubtext2)
```

Какие операторы использовались?

| Оператор | Что он делает |
|:-:|:-:|
| \s | любой пробельный символ |
| {n,} | не менее n повторений |

### 2.1.5 Напишите регулярное выражение, которое удаляет идущие подряд повторы. Одно слово из группы должно остаться.

```python
subtext3 = 'Футбольный клуб клуб Арсенал выигрывал выигрывал английскую премьер премьер лигу в последний раз раз раз раз в 2004 году году году. Сейчас команда находится находится находится в стадии стагнации. В ближайшем будущем будущем чемпионство им им им им им им не светит.'
```

```python
print(subtext3)
```

```python
newsubtext3 = re.sub(r'\b([^\W\d_]+)(\s+\1)+\b', r'\1', subtext3)
print(newsubtext3)
```

Какие операторы использовались?

| Оператор | Что он делает |
|:-:|:-:|
| \W | все, кроме букв и цифр |
| \d | любая цифра  [0-9] |
| () | группирует выражение и возвращает найденный текст |
| [^ ] | любой символ, кроме тех, что в скобках |
| + | 1 и более вхождений шаблона, стоящего перед + |
| \s | любой пробельный символ |
| \1 | заменяет найденное совпадение единичным экземпляром найденного слова |
| \b | граница слова |

### 2.1.6 Напишите регулярное выражение, которое определяет, что подстрока является адресом электронной почты.

```python
subtext4 = 'alelakiza@yandex.ru, lakiza@yaru, 21341cqdaa@gmail.com, blabla@mailru'

re.findall(r'\S+@\w+\.\w+', subtext4)
```

Какие операторы использовал?

| Оператор | Что он делает |
|:-:|:-:|
| + | 1 и более вхождений шаблона, стоящего перед + |
| \S | любой непробельный символ |
| \w | любая цифра или буква |
